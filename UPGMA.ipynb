```python
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import matplotlib.pyplot as plt

# 读取距离矩阵
def read_distance_matrix(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        distance_matrix = [[float(val) for val in line.strip().split()] for line in lines]
    return np.array(distance_matrix)

# 构建演化树
def build_evolutionary_tree(distance_matrix):
    # 使用UPGMA算法构建树
    linkage_matrix = linkage(distance_matrix, method='average')  
    # 绘制树状图
    dendrogram(linkage_matrix, orientation='top')
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.tick_params(axis='y', which='both', left=False, right=False, labelleft=False)
    plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=True)
    plt.savefig('evolutionary_tree.png' , dpi=300)
    plt.show()

# 主函数
def main():
    # 读取距离矩阵
    distance_matrix = read_distance_matrix('/data/home/liuzw/NR/pdb/structural/horizontal_matrices.txt')
    # 构建演化树
    build_evolutionary_tree(distance_matrix)

if __name__ == "__main__":
    main()
```
